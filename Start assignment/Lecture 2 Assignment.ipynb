### Question 1:
Step 1: Express the expected value \( E[y] \) in terms of \( P(y) \):

$E[y] = \sum_{y} y \cdot P(y) $

Step 2: Substitute the given joint probability distribution for \( P(y) \):

$E[y] = \sum_{y} y \cdot \left(\frac{n!}{y_1!y_2!y_3!...y_k!} \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k}\right) $

Step 3: Expand the sum:

$ E[y] = \sum_{y} \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (y_1 \cdot p_1) \cdot (y_2 \cdot p_2) \cdot (y_3 \cdot p_3) \ldots (y_k \cdot p_k) $

Step 4: Rearrange the terms:

$ E[y] =  \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot \sum_{y} (y_1 \cdot p_1) \cdot (y_2 \cdot p_2) \cdot (y_3 \cdot p_3) \ldots (y_k \cdot p_k) $

Step 5: Observe that the sum is a product of sums:

$E[y] =  \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot  \left( \sum_{y_1} \,y_1 \cdot p_1 \right) \cdot \left( \sum_{y_2} \, y_2 \cdot p_2 \right) \cdot \left( \sum_{y_3} \, y_3 \cdot p_3 \right) \ldots \left( \sum_{y_k} \, y_k \cdot p_k \right) $

Step 6: Each individual sum is simply the expected value of a binomial distribution with parameters \( n \) and \( p_i \):

$ E[Y_i] = n \cdot p_i $

Step 7: Substitute the expected values of each \( Y_i \):

$E[y] = \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot (n \cdot p_1) \cdot (n \cdot p_2) \cdot (n \cdot p_3) \ldots  (n \cdot p_k)$

Step 8: Notice that \( n! \) and \( y_1!y_2!y_3!...y_k! \) cancel out:

$E[y] = n \cdot p_1 \cdot n \cdot p_2 \cdot n \cdot p_3 \ldots n \cdot p_k $

We simplify the expression by combining terms and recognizing that the coefficient $(y_1!y_2!y_3!...y_k!) $ can be simplified to$ n! $, the number of permutations of outcomes.

Step 9: Simplify:

$E[y] = n  (p_1 \cdot  p_2 \cdot  p_3 \ldots  p_k) $

Step 10: Recall that \( p_1 + p_2 + p_3 + \ldots + p_k = 1 \):

$ E[y] = n \cdot 1 $

$ E[y] = n \cdot p_i $

To conclude, the expected value $E[y]$ equals $\Sigma y \cdot P(y) $, which is equal to $n \cdot p_i$.

Task 2: Consider the following experiment:

2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

Question 1: From the following steps shown above, derive the probability distribution of the experiment.

Question 2: Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.

Question 3: Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?

Question 4: Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.


$\textbf{Answers for task 2}$
-

### Question 1
The probability distribution of the experiment can be derived using the binomial distribution formula: $P(Y = y) = \binom{n}{y} \cdot p^y \cdot q^{n-y}$
- $P(Y = y)$ is the probability of observing exactly $y$ successes,
- $n$ is the number of trials,
- $p$ is the probability of success on a single trial,
- $q = (1 - p)$ is the probability of failure on a single trial.


### Question 2

Step 1: Define the probability mass function (PMF) of the binomial distribution:

$ P(Y = y) = \binom{n}{y} \cdot p^y \cdot q^{n-y} $

Where:
- $ n $ is the number of trials,
- $ y $ is the number of successes,
- $ p $ is the probability of success on a single trial,
- $ q = (1 - p) $ is the probability of failure on a single trial.

Step 2: Express the expectation $ E[Y] $ as a sum of $ y \cdot P(Y = y) $ over all possible values of $ y $:

$E[Y] = \sum_{y=0}^{n} y \cdot P(Y = y) $

Step 3: Substitute the expression for $P(Y = y)$ into the expectation formula:

$ E[Y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} \cdot p^y \cdot q^{n-y} $

Step 4: Expand the binomial coefficient $ \binom{n}{y}$ using its definition:

$ E[Y] = \sum_{y=0}^{n} y \cdot \frac{n!}{y!(n-y)!} \cdot p^y \cdot q^{n-y}$

Step 5: Rearrange terms:

$E[Y] = \sum_{y=0}^{n} \frac{n!}{(y-1)!(n-y)!} \cdot p^y \cdot q^{n-y} $

Step 6: Rewrite the sum starting from $ y = 1 $, since the term $ y = 0 $ contributes zero to the sum:

$ E[Y] = \sum_{y=1}^{n} \frac{n!}{(y-1)!(n-y)!} \cdot p^y \cdot q^{n-y} $

Step 7: Factor out $ np $ from the sum:

$E[Y] = np \cdot \sum_{y=1}^{n} \frac{(n-1)!}{(y-1)!(n-1-(y-1))!} \cdot p^{y-1} \cdot q^{n-1-(y-1)} $

Step 8: Rewrite the sum using the binomial coefficient $\binom{n-1}{y-1} $:

$E[Y] = np \cdot \sum_{y=0}^{n-1} \frac{(n-1)!}{y!(n-1-y)!} \cdot p^{y} \cdot q^{n-1-y}$

Step 9: Note that $\sum_{y=0}^{n-1} \binom{n-1}{y} \cdot p^{y} \cdot q^{n-1-y} $ is simply $(p + q)^{n-1} $, which equals 1. 

Step 10: Simplify:

$ E[Y] = np \cdot 1^{n-1} = np $

So, we've shown that the expectation $ E[Y] $ of this probability distribution is indeed $E[Y] = \Sigma y \cdot P(y) = np $.

### Question 3:

Step 1: Define the experiment:
- The experiment involves administering a new medication to individuals with a certain illness.
- The probability of an individual recovering with the medication is given as 30%, which implies $ p = 0.30 $.
- The number of trials (individuals receiving the medication) is $ n = 10 $.
- The random variable of interest is $ Y $, representing the number of successes (individuals who recover) out of $ n $ trials.

Step 2: Determine the probability distribution:
- Since the experiment follows a binomial distribution, the probability mass function (PMF) is given by:
  $ P(Y = y) = \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y} $
  where $ \binom{n}{y} $ is the binomial coefficient, $ p $ is the probability of success, and $ (1-p) $ is the probability of failure.

Step 3: Calculate the probability of at least nine individuals recovering:
- We want to find $P(Y \geq 9) $, the probability that at least nine out of ten individuals recover.
- This can be calculated by summing the probabilities of exactly nine successes and exactly ten successes:
  $ P(Y \geq 9) = P(Y = 9) + P(Y = 10) $

Step 4: Substitute values into the binomial PMF and calculate probabilities:
- For $ P(Y = 9) $:
  $ P(Y = 9) = \binom{10}{9} \cdot 0.30^9 \cdot (1-0.30)^{10-9} $
- For $ P(Y = 10) $:
  $ P(Y = 10) = \binom{10}{10} \cdot 0.30^{10} \cdot (1-0.30)^{10-10} $


Step 5: Substitute values into the binomial PMF and calculate probabilities:
- For $P(Y = 9) $:
  $ P(Y = 9) = \binom{10}{9} \cdot 0.30^9 \cdot (1-0.30)^{10-9} $
 $P(Y = 9) = 10 \cdot 0.30^9 \cdot 0.70^{1} $
 $ P(Y = 9) \approx 0.121 $

- For $P(Y = 10) $:
 $ P(Y = 10) = \binom{10}{10} \cdot 0.30^{10} \cdot (1-0.30)^{10-10} $
  $ P(Y = 10) = 0.30^{10} $
  $P(Y = 10) \approx 0.028 $

Step 6: Calculate the total probability and percentage likelihood:
- $P(Y \geq 9) = P(Y = 9) + P(Y = 10)$
- $ P(Y \geq 9) \approx 0.121 + 0.028 $
- $ P(Y \geq 9) \approx 0.149$

To convert this probability to a percentage, we multiply by 100:
- Percentage likelihood of at least nine out of ten individuals recovering: $ 0.149 \times 100\% $
- Percentage likelihood $ \approx 14.9\% $

So, there's approximately a 14.9% chance that at least nine out of ten individuals receiving the medication will recover, even if the medication is completely ineffective.


### Question 4:

-

### Question 1:

Step 1: Use the formula for the expected value of a discrete random variable:
$ E[Y] = \sum_{y=0}^{\infty} y \cdot p(y)$

Step 2: Substitute the given probability distribution $ p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $ into the formula:
$E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} $
  
Step 3: Recognize that the sum simplifies to the expected value of a Poisson distribution:
$ E[Y] = \lambda $

 The expected value $ E[Y] $ of the given probability distribution is simply equal to its parameter $\lambda $.

 ### Question 2:

Step 1: Define the problem:

Step 2: Identify the parameters:
